```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("top_keywords_sv.csv")
```

```python
df.sort_values("Search volume").head(10)
```

```python
df.sort_values("Search volume", ascending=False).head(10)
```

```python
frequencies = {row['Keyword']: row['Search volume'] for _, row in df.iterrows()}
```

```python
keywords_cloud = WordCloud(width=800, 
                           height=400, 
                           max_words=100, 
                           background_color="white", 
                           min_font_size=10,
                           relative_scaling=1,
                          prefer_horizontal=True).generate_from_frequencies(frequencies)
```

```python
plt.axis("off")
plt.imshow(keywords_cloud, interpolation="bilinear")
plt.show()
```
